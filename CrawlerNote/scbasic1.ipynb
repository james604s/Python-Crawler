```python
import json
src = "https://data.taipei/opendata/datalist/apiAccess?scope=resourceAquire&rid=296acfa2-5d93-4706-ad58-e83cc951863c"
with request.urlopen(src) as response:
    data = json.load(response) 
print(data)
```

```python
#將公司列表列出來
#data是json全部的程式檔案、對result及底下的results有興趣
clist = data["result"]["results"]
print(clist)

```

```python
with open("data.csv", "w", encoding="utf-8") as file: #打開一檔案寫入並用utf-8編碼
    for company in clist:
        file.write(company["公司名稱"] + "\n") #檔案寫入公司名稱並換行
```
